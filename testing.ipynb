```python
X, Y = np.meshgrid(np.arange(0, 2 * np.pi, .2), np.arange(0, 2 * np.pi, .2))
U = np.cos(X)
V = np.sin(Y)

# Normalize the arrows:
U = U / np.sqrt(U**2 + V**2);
V = V / np.sqrt(U**2 + V**2);


plt.figure()
plt.title('Normalized arrows')
Q = plt.quiver(X, Y, U, V, units='width')
qk = plt.quiverkey(Q, 0.9, 0.9, 2, r'$2 \frac{m}{s}$', labelpos='E',
                   coordinates='figure')
```

```python
n = 8
X, Y = np.mgrid[0:n, 0:n]
T = np.arctan2(Y - n / 2., X - n/2.)
R = 10 + np.sqrt((Y - n / 2.0) ** 2 + (X - n / 2.0) ** 2)
U, V = R * np.cos(T), R * np.sin(T)

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.quiver(X, Y, U, V, R, alpha=.5)
plt.quiver(X, Y, U, V, edgecolor='k', facecolor='None', linewidth=.5)

plt.xlim(-1, n)
plt.xticks(())
plt.ylim(-1, n)
plt.yticks(())

plt.show()
```

```python
from __future__ import division

%matplotlib inline
```

```python
def elec_field(x, y, x_q, y_q, q):
    R = np.sqrt((x - x_q)**2 + (y - y_q)**2) + 1e-12
    Ex = q*(x - x_q)/R**2
    Ey = q*(y - y_q)/R**2
    return Ex, Ey
```

```python
xi = np.linspace(-5, 5, 30)
yi = np.linspace(-5, 5, 30)
x, y = np.meshgrid(xi, yi)
    
Ex, Ey = elec_field(x, y, 0, 0, 1)
Emag = np.sqrt(Ex**2 + Ey**2)
dir_x = Ex/Emag
dir_y = Ey/Emag
plt.quiver(x, y, dir_x, dir_y)
plt.scatter(0, 0, marker='o')
plt.show()
```
